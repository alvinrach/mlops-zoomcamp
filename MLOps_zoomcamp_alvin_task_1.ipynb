```python
from sklearn.linear_model import LinearRegression
```

```python
y = (data.tpep_dropoff_datetime-data.tpep_pickup_datetime).dt.seconds/60
```

```python
X = data[['PULocationID', 'DOLocationID']].astype('str')
X = pd.get_dummies(X)
```

```python
lr = LinearRegression()
lr.fit(X,y)
```

```python
y_pred = lr.predict(X)
```

```python
from sklearn.metrics import mean_squared_error
mean_squared_error(y, y_pred, squared=False)
```

```python
data_feb = pd.read_parquet('Downloads/yellow_tripdata_2022-02.parquet')
X_test = data_feb[['PULocationID', 'DOLocationID']].astype('str')
X_test = pd.get_dummies(X_test)
y_true = (data_feb.tpep_dropoff_datetime-data_feb.tpep_pickup_datetime).dt.seconds/60
y_pred = lr.predict(X_test)
mean_squared_error(y_true, y_pred, squared=False)
```
